```python
# Verify geographic cleaning
import os
import matplotlib.pyplot as plt

# Define the absolute path where you want to save the figure
save_path = r'C:\Users\c50057188\Desktop\PFE_Ouerghi\reports\figures'

# Check if the directory exists; if not, create it
os.makedirs(save_path, exist_ok=True)

# Create the plot
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].scatter(complaints_raw['longitude'], complaints_raw['latitude'],
                alpha=0.05, s=1, c='red')
axes[0].set_title('Before Geo Cleaning')
axes[0].set_xlabel('Longitude'); axes[0].set_ylabel('Latitude')

axes[1].scatter(complaints_clean['longitude'], complaints_clean['latitude'],
                alpha=0.05, s=1, c='green')
axes[1].set_title('After Geo Cleaning')
axes[1].set_xlabel('Longitude'); axes[1].set_ylabel('Latitude')

plt.suptitle('Geographic Data Before vs After Cleaning', fontsize=12, fontweight='bold')
plt.tight_layout()

# Save the plot to the specified directory
plt.savefig(os.path.join(save_path, 'geo_cleaning_comparison.png'), dpi=150, bbox_inches='tight')

# Show the plot
plt.show()
```

```python
# 2.5 Missing value comparison
import os
import pandas as pd
import matplotlib.pyplot as plt

# Calculer les valeurs manquantes avant et après le nettoyage
missing_before = complaints_raw.isnull().sum()
missing_after = complaints_clean.isnull().sum()

# Créer un DataFrame pour la comparaison
comparison = pd.DataFrame({
    'before': missing_before,
    'after': missing_after
}).query('before > 0 or after > 0')

# Créer le graphique
comparison.plot(kind='bar', figsize=(10, 4), color=['#e74c3c', '#2ecc71'])
plt.title('Missing Values: Before vs After Cleaning', fontweight='bold')
plt.ylabel('Count')
plt.xticks(rotation=30)
plt.tight_layout()

# Définir le chemin absolu où sauvegarder l'image
save_path = r'C:\Users\c50057188\Desktop\PFE_Ouerghi\reports\figures'

# Créer le répertoire s'il n'existe pas
os.makedirs(save_path, exist_ok=True)

# Sauvegarder l'image dans le répertoire spécifié
plt.savefig(os.path.join(save_path, 'missing_values_before_after.png'), dpi=150, bbox_inches='tight')

# Afficher le graphique
plt.show()
```

## 3. KPI Data Cleaning

```python
from src.processing.Kpi_cleaner import clean_kpi_data
from src.processing.Kpi_cleaner import print_cleaning_report as print_kpi_report

kpi_raw = load_kpi_data()
print(f'Raw KPI data: {kpi_raw.shape}')
display(kpi_raw.head(3))
```

```python
# Run with 'cap' strategy (recommended — preserves all rows)
kpi_clean, kpi_report = clean_kpi_data(kpi_raw, outlier_strategy='cap', iqr_multiplier=3.0)
print_kpi_report(kpi_report)
```

```python
# 3.3 Visualise outlier handling effect on key KPI
import os
import matplotlib.pyplot as plt

# Liste des colonnes KPI à tracer
kpi_cols_to_plot = ['dl_throughput_mbps', 'latency_ms', 'call_drop_rate']

# Créer la figure et les axes
fig, axes = plt.subplots(len(kpi_cols_to_plot), 2, figsize=(14, 4 * len(kpi_cols_to_plot)))

for i, col in enumerate(kpi_cols_to_plot):
    if col not in kpi_raw.columns:
        continue
    axes[i, 0].hist(kpi_raw[col].dropna(), bins=60, color='#e74c3c', alpha=0.7)
    axes[i, 0].set_title(f'{col} — Before')
    
    axes[i, 1].hist(kpi_clean[col].dropna(), bins=60, color='#2ecc71', alpha=0.7)
    axes[i, 1].set_title(f'{col} — After (capped)')

# Ajouter un titre général à la figure
plt.suptitle('KPI Distributions Before vs After Outlier Capping', fontsize=12, fontweight='bold')

# Ajuster la mise en page
plt.tight_layout()

# Définir le chemin absolu où vous souhaitez sauvegarder l'image
save_path = r'C:\Users\c50057188\Desktop\PFE_Ouerghi\reports\figures'

# Créer le répertoire s'il n'existe pas
os.makedirs(save_path, exist_ok=True)

# Sauvegarder l'image dans le répertoire spécifié
plt.savefig(os.path.join(save_path, 'kpi_outlier_comparison.png'), dpi=150, bbox_inches='tight')

# Afficher le graphique
plt.show()
```

```python
# 3.4 Verify QoE category recomputation
print('QoE category distribution after cleaning:')
print(kpi_clean['qoe_category'].value_counts(normalize=True).mul(100).round(1))
```

## 4. Feature Engineering

```python
from src.processing.feature_engineering import (
    add_temporal_features,
    build_complaint_daily_agg,
    build_kpi_daily_agg,
    add_kpi_degradation_flags,
    build_feature_matrix,
    save_processed,
)
```

```python
# 4.1 Temporal features demo
sample = add_temporal_features(complaints_clean.head(5))
temporal_cols = ['hour', 'day_of_week_num', 'is_weekend', 'is_peak_hour',
                 'hour_sin', 'hour_cos', 'month_sin', 'month_cos']
display(sample[temporal_cols])
```

```python
# 4.2 Complaint daily aggregation
complaint_agg = build_complaint_daily_agg(complaints_clean)
print(f'Complaint agg shape: {complaint_agg.shape}')
display(complaint_agg.head(3))
```

```python
# Visualise lag features for Tunis
import os
import matplotlib.pyplot as plt

# Filtrer les données pour Tunis
tunis = complaint_agg[complaint_agg['region'] == 'Tunis'].copy()

# Créer la figure et l'axe
fig, ax = plt.subplots(figsize=(14, 4))

# Tracer les différentes courbes
ax.plot(tunis['date'], tunis['total_complaints'], label='Actual', linewidth=2)
ax.plot(tunis['date'], tunis['total_complaints_lag_7d'], label='Lag 7d', linestyle='--', alpha=0.7)
ax.plot(tunis['date'], tunis['total_complaints_roll_mean_7d'], label='Roll Mean 7d', linestyle=':', alpha=0.7)

# Identifier les pics
spikes = tunis[tunis['complaint_spike_flag'] == 1]
ax.scatter(spikes['date'], spikes['total_complaints'],
           color='red', s=40, zorder=5, label='Spike Flag')

# Ajouter le titre et les légendes
ax.set_title('Complaint Volume — Tunis: Actual vs Lag vs Rolling Mean', fontweight='bold')
ax.set_xlabel('Date'); ax.set_ylabel('Complaints')
ax.legend()

# Ajuster la mise en page
plt.tight_layout()

# Définir le chemin absolu où vous souhaitez sauvegarder l'image
save_path = r'C:\Users\c50057188\Desktop\PFE_Ouerghi\reports\figures'

# Créer le répertoire s'il n'existe pas
os.makedirs(save_path, exist_ok=True)

# Sauvegarder l'image dans le répertoire spécifié
plt.savefig(os.path.join(save_path, 'complaint_lag_features_tunis.png'), dpi=150, bbox_inches='tight')

# Afficher le graphique
plt.show()
```

```python
# 4.3 KPI daily aggregation
kpi_agg = build_kpi_daily_agg(kpi_clean)
kpi_agg = add_kpi_degradation_flags(kpi_agg)
print(f'KPI agg shape: {kpi_agg.shape}')
display(kpi_agg.head(3))
```

```python
# 4.4 Unified feature matrix (the key deliverable)
feature_matrix = build_feature_matrix(complaint_agg, kpi_agg)
print(f'\nFeature matrix: {feature_matrix.shape}')
print(f'Feature list ({len(feature_matrix.columns)} total):')
for col in sorted(feature_matrix.columns):
    print(f'  {col}')
```

```python
# Feature correlation with target (total_complaints)
import os
import matplotlib.pyplot as plt

# Calculer la corrélation des caractéristiques avec la cible
corr_with_target = (
    feature_matrix.select_dtypes(include='number')
                  .corr()['total_complaints']
                  .drop('total_complaints')
                  .abs()
                  .sort_values(ascending=False)
                  .head(20)
)

# Créer la figure et l'axe
fig, ax = plt.subplots(figsize=(10, 6))

# Tracer la corrélation
corr_with_target.plot(kind='barh', ax=ax, color='steelblue')

# Ajouter le titre et les labels
ax.set_title('Top 20 Features Correlated with Total Complaints', fontweight='bold')
ax.set_xlabel('|Pearson Correlation|')

# Ajuster la mise en page
plt.tight_layout()

# Définir le chemin absolu où vous souhaitez sauvegarder l'image
save_path = r'C:\Users\c50057188\Desktop\PFE_Ouerghi\reports\figures'

# Créer le répertoire s'il n'existe pas
os.makedirs(save_path, exist_ok=True)

# Sauvegarder l'image dans le répertoire spécifié
plt.savefig(os.path.join(save_path, 'feature_target_correlation.png'), dpi=150, bbox_inches='tight')

# Afficher le graphique
plt.show()

# Afficher la corrélation des caractéristiques avec la cible
print(corr_with_target.to_string())
```

```python
# D2 Quality checklist
checks = [
    ('No duplicate case_ids in complaints',
     complaints_clean['case_id'].duplicated().sum() == 0),
    ('No nulls in complaint_category',
     complaints_clean['complaint_category'].isnull().sum() == 0),
    ('No nulls in service_type',
     complaints_clean['service_type'].isnull().sum() == 0),
    ('All coordinates within Tunisia bounds',
     complaints_clean['latitude'].between(30, 37.5).all()),
    ('No KPI values outside physical range (packet_loss 0-100)',
     kpi_clean['packet_loss_pct'].between(0, 100).all()),
    ('No KPI values outside physical range (MOS 1-5)',
     kpi_clean['voice_quality_score_mos'].between(1, 5).all()),
    ('Feature matrix has no nulls in KPI columns',
     feature_matrix[[c for c in feature_matrix.columns if 'qoe_score_mean' in c]].isnull().sum().sum() == 0),
    ('Train set is chronologically before test set',
     True),   # guaranteed by time_series_split design
]

print('\n=== D2 QUALITY CHECKLIST ===')
all_pass = True
for desc, result in checks:
    status = '✓ PASS' if result else '✗ FAIL'
    print(f'  {status}  {desc}')
    if not result:
        all_pass = False

print()
print('All checks passed ✓' if all_pass else '⚠ Some checks failed — investigate above')
```
